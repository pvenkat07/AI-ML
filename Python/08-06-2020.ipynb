## Merging : inner,outer,left,right

```python
df1 = pd.DataFrame({"branch":['hyd','blr','vizag','pune'],"total":[90,96,98,92]})
df2 = pd.DataFrame({"branch":['blr','hyd','vizag','chennai'],"distance_from_city":[40,60,12,20]})
df1
```

```python
df2
```

```python
pd.merge(df1,df2,on="branch",how='inner')
```

```python
pd.merge(df1,df2,on="branch",how='outer',indicator=True)
```

```python
pd.merge(df1,df2,on="branch",how='left',indicator=True)
```

```python
pd.merge(df1,df2,on="branch",how='right',indicator=True)
```

## Simple Arithmetic operations on two or more dataframes
- add()
- sub()
- mul()
- div()
- floordiv()
- mod()
- pow()

```python
IPL_2018 = pd.DataFrame({'IPL Team': ['CSK', 'SRH', 'KKR', 'RR', 'MI', 'RCB', 'KXIP', 'DD'],
                         'Matches Played': [16, 17, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [11, 10, 9, 7, 6, 6, 6, 5]}
                       )
IPL_2018.set_index('IPL Team', inplace = True)
IPL_2017 = pd.DataFrame({'IPL Team': ['MI', 'RPS', 'KKR', 'SRH', 'KXIP', 'DD', 'GL', 'RCB'],
                         'Matches Played': [17, 16, 16, 15, 14, 14, 14, 14],
                         'Matches Won': [12, 10, 9, 8, 7, 6, 4, 3]}
                       )
IPL_2017.set_index('IPL Team', inplace = True)
IPL_2017
```

```python
IPL_2018
```

```python
Total = IPL_2018 +IPL_2017
Total
```

```python
Total = IPL_2018.add(IPL_2017,fill_value=0)
Total
```

```python
IPL_2018.mul(IPL_2017,fill_value=0)
```

```python
IPL_2018.sub(IPL_2017,fill_value=0)
```

```python
IPL_2018.div(IPL_2017,fill_value=0)
```

```python
IPL_2018.mod(IPL_2017,fill_value=0)
```

## groupby() : split,apply,combine:

```python
data = {"name":['a','b','c','d','e','f'],
       "branch":['hyd','vizag','blr','hyd','vizag','blr'],
       "marks":[90,91,92,93,94,95]}
df = pd.DataFrame(data)
df
```

```python
d = df.groupby('branch')
d
```

```python
for branch,branch_df in d:
    print(branch) 
    print(branch_df) 
```

```python
d.get_group('hyd')["marks"].sum()
```

```python
d.get_group('hyd')["marks"].mean()
```
