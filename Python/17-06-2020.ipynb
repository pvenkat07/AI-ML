## Feature Scaling: Not manditoryfor all algorithms

```python
sample = pd.DataFrame({'w.e': [2,4,5,6,7,8],'Age': [28,25,32,45,33,32],
                      'salary' : [3,4,5,6,7,8],'Distance' : [10,12,23,32,15,17],
                     'expectedsalary': [4,5,6,7,8,9]})
sample
```

```python
from sklearn.preprocessing import scale
scaled_data = scale(sample)
pd.DataFrame(scaled_data)
```

```python
sample_data = sample.copy()
sample_data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sample_data = sc.fit_transform(sample_data)
pd.DataFrame(sample_data)
```

```python
sample_d= sample.copy()
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
mm.fit(sample_d)
sample_d = mm.transform(sample_d)
pd.DataFrame(sample_d)
```
